ðŸ”¥ PROJECT 1: Parallel Monte Carlo Simulation (Ï€ Estimation)
ðŸ“Œ Concept

Monte Carlo simulations are independent tasks, ideal for parallel execution.

```r
cl <- makeCluster(detectCores() - 1)
registerDoParallel(cl)

n <- 1e6

inside_circle <- foreach(i = 1:n, .combine = "+") %dopar% {
  x <- runif(1)
  y <- runif(1)
  if ((x^2 + y^2) <= 1) 1 else 0
}

stopCluster(cl)

pi_estimate <- 4 * inside_circle / n
pi_estimate
```

ðŸ”¥ PROJECT 2: Parallel Time Series ARIMA Modeling
ðŸ“Œ Concept

Fit ARIMA models on multiple datasets simultaneously.

```r
time_series_list <- list(
  AirPassengers,
  nottem,
  Nile
)

fit_arima <- function(ts_data) {
  auto.arima(ts_data)
}

cl <- makeCluster(detectCores() - 1)
models <- parLapply(cl, time_series_list, fit_arima)
stopCluster(cl)

models
```


ðŸ”¥ PROJECT 3: Parallel Bootstrap Confidence Interval
ðŸ“Œ Concept
Bootstrap resampling is computationally expensive â†’ parallelism improves speed.



```r
data <- rnorm(1000)

boot_mean <- function(data, index) {
  mean(data[index])
}

boot_results <- boot(
  data,
  statistic = boot_mean,
  R = 5000,
  parallel = "snow",
  ncpus = detectCores() - 1
)

boot_results
```

ðŸ”¥ PROJECT 4: Parallel Text Mining
ðŸ“Œ Concept

Each document is processed independently.

```r
texts <- c(
  "Parallel computing improves performance",
  "R programming supports parallel execution",
  "Big data analytics using R"
)

process_text <- function(text) {
  words <- strsplit(tolower(text), "\\s+")[[1]]
  table(words)
}

cl <- makeCluster(detectCores() - 1)
word_freq <- parLapply(cl, texts, process_text)
stopCluster(cl)

word_freq
```
